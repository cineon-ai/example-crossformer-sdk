```python
import os
from pathlib import Path

import numpy as np
import torch
from cineon_format import CineonData

from example_crossformer_sdk import CrossformerModel
```

```python
# Go up a directory if we're in the notebooks dir
_cwd = Path().cwd()
if os.path.basename(_cwd) == "notebooks":
    os.chdir(Path().cwd().parents[0])
```

```python
# Load the model
model = CrossformerModel("model")
```

```python
# Load the data
data = CineonData.from_csv("data/example.csv")
```

```python
# Preprocess the data and run it through the model
features, mask = model.preprocess([data])
logits = model.forward(features, mask)
print(f"features.shape: {features.shape}")
print(f"logits: {logits}")
```

```python
# The entire model prediction pipeline can also be run in one step using the `run` method,
# The run method handles both preprocessing and forward pass.
results = model.run([data])
print(f"results: {results}\n")
print(f"Logits: {results.logits}\n")

# Get prediction probabilities
probabilities = torch.softmax(torch.tensor(results.logits), dim=1)
print(f"probabilities: {probabilities}\n")

# Get predicted classes
predicted_classes = np.argmax(probabilities.numpy(), axis=1)
print(f"predicted_classes: {predicted_classes}")
```
